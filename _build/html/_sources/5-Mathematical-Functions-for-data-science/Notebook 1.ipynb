# Algebraic functions

## Linear functions

Tiene la forma de $$f(x)=mx + b$$ donde $m$ y $b$  $\in R$. 

$m$ puede ser calculada por: $$m=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

y $b$ es el punto de corte con el eje $y$. Su dominio es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = (-\infty, \infty)$



### coding a linear function

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100

#slope
m = -1

#intercept
b = 3

#let's define the function
def f(x):
  return (m*x + b)

#domain
x = np.linspace(-10,10, num=N)

#range
y = f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
```

## Polinomic functions

Tiene la forma de $$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{1}$$

a una funci√≥n que tiene esta forma se le llama polinomio de grado $n$. A los elementos $a$ los llamaremos coeficientes donde $a \in R$. 

**Por ejemplo:**

$$P(x)= 2x^{7} - x^{4} + 3x^{2} + 4$$

que es un polinomio de grado 7.


```python
def f2(x):
    return(2*(x**7) - x**4 + 3*(x**2) + 4)

y2 = f2(x)

fig, ax2 = plt.subplots()
ax2.plot(x,y2)
ax2.grid()
```

## Power Function

Hay unas funciones que son un caso particular de las funciones polin√≥micas que son las funciones potencia, las cuales tienen la forma:


$$f(x)= x^{a}, a \in R$$ 


**Por ejemplo:**

$$f(x)= x^{2}$$

El dominio de $f(x)=x^{2}$ es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = [0, \infty)$


### Coding a power function

```python
def f3(x):
  return 7**x

y3 = f3(x)

fig, ax3 = plt.subplots()
ax3.plot(x,y3)
ax3.grid()
```

## Exponential function

Tienen la forma de $$f(x)=a^x$$ donde la base $a$ es una constante positiva. Un gran ejemplo de una funci√≥n exponencial es usando la base como el n√∫mero de euler:

$$f(x)=e^x$$

```python
def f(x):
  return np.exp(x)    #euler

y=f(x)

plt.plot(x,y)
```

## Logarithmic functions

El logaritmo est√° definido por la **relaci√≥n**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$ 

donde: 



*   $b$ es la base.
*   $n$ es el exponente al que est√° elevado la base.
*   $x$ es el resultado de elevar la base $b$ al exponente $n$

**Ejemplo:**

Teniendo b=2 y n=8, entonces:

$$2^8=256$$

Por lo que $x=256$. Calculando el logaritmo base 2 de $x$ es:

$$log_{2}(256) = 8$$

```python
def f(x):
    return np.log2(x)

#domain
x = np.linspace(0, 10, 1000)

plt.plot(x, f(x))
plt.show()
```

## Heavyside function

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la funci√≥n escal√≥n de Heaviside: 

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
x = np.linspace(-10, 10, 1000)

def f(x):
    
    Y = np.zeros(len(x))
    
    for index, x in enumerate(x):
        if x >= 0:
            Y[index] = 1
    return Y

plt.plot(x, f(x))
plt.show
```

# How to manipulate functions

## Functions displacement

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```python
N = 1000

def f(x):
  return x**2;

c = 4

x = np.linspace(-10,10, num=N)

y = f(x + c)     #note that this will displace y 4 units to the right

fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x, f(x))     #original cuadratic function
ax[0].grid()
ax[0].axhline(y=0, color='r')
ax[0].axvline(x=0, color='r')
ax[0].set_title("Original chart")

ax[1].plot(x,y)         #function with transformation
ax[1].grid()
ax[1].axhline(y=0, color='r')
ax[1].axvline(x=0, color='r')
ax[1].set_title("Displaced chart")
```

## Functions elongations & compressions

Siendo $c$ una constante mayor que cero, entonces la gr√°fica:

- $y=c \cdot f(x)$ alarga la gr√°fica verticalmente en un factor de $c$.
- $y= \frac{1}{c} \cdot f(x)$ comprime la gr√°fica verticalmente en un factor de $c$.
- $y=f(c \cdot x)$ comprime la gr√°fica horizontelmente en un factor de $c$.
- $y= f(\frac{1}{c} \cdot x )$ alarga la gr√°fica horizontelmente en un factor de $c$.

```python
N = 1000

def f(x):
  return np.sin(x);

c = 2

x = np.linspace(-15,15, num=N)

y = f((1/c)*x)     #note that this will elonge the chart in a proportion of 2


fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x, f(x))     #original sinusoidal function
ax[0].grid()
ax[0].axhline(y=0, color='r')
ax[0].axvline(x=0, color='r')
ax[0].set_title("Original chart")

ax[1].plot(x,y)         #function with transformation (elonged horizontally)
ax[1].grid()
ax[1].axhline(y=0, color='r')
ax[1].axvline(x=0, color='r')
ax[1].set_title("Elonged chart")
```

## Function reflection

- $y=-f(x)$ refleja la gr√°fica respecto al eje x.
- $y=f(-x)$ refleja la gr√°fica respecto al eje y.

```python
N = 1000

def f(x):
  return x**3;

x = np.linspace(-10,10, num=N)

y = f(-x)      #this will reflect the function about the y axis

fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x, f(x))     #original cubic function
ax[0].grid()
ax[0].axhline(y=0, color='r')
ax[0].axvline(x=0, color='r')
ax[0].set_title("Original chart")

ax[1].plot(x,y)         #function with transformation (reflected about the y axis)
ax[1].grid()
ax[1].axhline(y=0, color='r')
ax[1].axvline(x=0, color='r')
ax[1].set_title("Reflected chart")
```

# Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-5,5, num=N)
```

## Linear function

$$y=mx+b$$

```python
def f(x):
  return x

plt.plot(x, f(x))
plt.grid()
```

## heavyside function

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    

N=1000


y = H(x)

plt.plot(x,y)
plt.grid()
```

## sigmoid function

$$f(x)=\frac{1}{1-e^{-x}}$$

```python
def f(x):
  return 1/(1 + np.exp(-x))
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## hyperbolic tangent function

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

```python
def f(x):
  return np.tanh(x)
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## ReLu function

$$R(x)=max(0,x)$$

```python
def f(x):
  return np.maximum(x,0)
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```
